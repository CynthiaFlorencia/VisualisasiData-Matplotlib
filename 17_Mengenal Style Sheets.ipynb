```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as seaborn
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

**Predefined Styles**

- Untuk melihat predefined style kita memanggil fungsi plt.style.available
- Untuk menggunakan predefined style kita menggunakan fungsi plt.style.use().

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

- Membuat 15 buat nilai random yang ditampung kedalam variabel x
- Membuat 2 buah plot
- Plot pertama akan dilakukan plotting variabel x terhadap nilai x juga
- Parameter 'bs-' pada plotting pertama berarti 'blue, square, -'
- Plot kedua akan dilakukan plotting variabel x terhadap nilai log x
- Parameter 'ro-' pada plotting kedua berarti 'red, dots, -'

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

**Menerapkan multiple style**

- Dalam matplotlib mungkin bagi kita untuk menerapkan lebih dari satu style
- Urutan style berpengaruh terhadap penerapan/ overlapping.**

```python
#Set Seaborn style
seaborn.set_style('whitegrid')  #Choose any Seaborn style you prefer

#Use the 'dark_background' style from Matplotlib
plt.style.use('dark_background')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

**Menerapkan temporary styling**

- Untuk menerapkan style sementara, kita bisa memasukkan style yang kita inginkan kedalam scope with.**

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')

plt.show()
```
