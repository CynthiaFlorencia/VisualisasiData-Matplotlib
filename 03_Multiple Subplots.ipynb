**Simple Line Plot**

- Membuat deret bilangan dengan fungsi arange(start, end, interval)
- End point bersifat exclusive (bersifat sebagai pembatas)
- Dalam trigonometri x dalam 2Ï€x berarti radian

```python
x = np.arange(0.0, 2.0, 0.01)
x
```

```python
s = np.sin(2 * np.pi* x)
s
```

```python
#Ciri utama 00 Style terdapat 2 variable penampung yaitu fig untuk figure, dan ax untuk axes 
fig, ax = plt.subplots() #00 Style 

#Sumbu x merepresentasikan variable x dan sumbu y merepresentasikan variable s
ax.plot(x,s)
ax.set(xlabel = 'nilaix',
       ylabel = 'nilai sin(x)',
       title = 'Visualisasi nilai sin')

#Fungsi grid berfungsi untuk menampilkan garis bantu 
ax.grid()

plt.show()
```

**Multiple Subplots**

```python
#linspace() memiliki parameter (start, end, jumlah data)
#Start dan end point pada linspace() bersifat inclusive
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

**Multiple Subplots dengan 00 Style**

- Contoh dibawah adalah multiple subplots, karena memiliki 1 figure dan 2 axes
- Gunakan plt.subplots(baris, kolom) untuk membuat 00 Style
- Buat 2 axes dengan tuple
- Pada kasus ini akan dibuat 2 baris dan 1 kolom
- Untuk menyertakan judul pada figure gunakan suptitle()
- Cara menerapkan LaTeX Style pada label matplotlib adalah dengan mengapit teks dengan dollar $
- Selanjutnya taruh underscore didepan teks yang ingin dikecilkan

```python
#Membuat objek subplot dengan 2 baris, 1 kolom
fig, (ax1, ax2) = plt.subplots(2,1)

#Menambahkan judul utama pada keseluruhan subplot
fig.suptitle('Multiple Subplots')

#Menampilkan data pada subplot pertama (ax1)
#Warna = merah, marker = 0-
ax1.plot(x1, y1, 'ro-')
#Label sumbu y dengan notasi LaTeX
ax1.set_ylabel('Nilai cos(x1)')

#Menampilkan data pada subplot kedua (ax2)
#Warna = hijau, marker = .-
ax2.plot(x2, y2, 'g-')
#Label sumbu y dengan notasi LaTeX
ax2.set_ylabel('Niali cos(x2)')

#Menambahkan label pada sumbu x subplot kedua (ax2)
ax2.set_xlabel('Nilai x')

#Menampilkan grafik
plt.show()
```

**Multiple Subplots dengan pyplot Style**

```python
#Parameter ketiga pada fungsi subplot() berguna untuk mengendalikan akses ke n

#Membuat subplot pertama (atas) 
plt.subplot(2, 1, 1)
#Warna = merah, marker = 0-
plt.plot(x1, y1, 'ro-')
#Label sumbu y dengan notasi LaTeX
plt.title('Multiple Subplots')
#Judul subplot pertama 
plt.ylabel('Nilai $cos(x_1)$')

#Membuat subplot kedua (bawah) 
plt.subplot(2, 1, 2)
#Warna = Hijau, Marker = .-
plt.plot(x2, y2, 'g--')
#Label sumbu y dengan notasi LaTeX
plt.ylabel('Nilai $cos(x_2)$')
#Label sumbu x subplot kedua 
plt.xlabel('Nilai $X$')

#Menampilkan grafik 
plt.show()
```
