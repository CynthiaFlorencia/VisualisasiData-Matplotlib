**Simple Bar Plot**

- Bar plot biasanya digunakan untuk melakukan plotting data yang bersifat categorical atau discrete
- plt.bar() digunakan untuk membuat bar plot
- Setidaknya ada dua parameter yang harus kita masukkan, yaitu: kategorical data dan dan numeric data
- kategorical data adalah parameter akan digunakan sebagai label
- Numeric data adalah parameter yang akan digunakan sebagai nilai

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori,data)
plt.grid()

plt.xlabel('kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

**Pengaturan Grid dan Color**

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

- Pada parameter plt.bar(), Color dan alpha digunakan untuk mengatur warna dan opacity bar
- Pada parameter plt.grid(), linestyle dan linewidth digunakan untuk mengatur bentuk dan ketebalan garis
- Pada parameter plt.grid(), Axis digunakan untuk mengatur garis agar dapat berasosiasi dengan sumbu
- Pada parameter plt.grid(), Alpha digunakan untuk mengatur opacity garis bantu

```python
#Membuat bar plot
plt.bar(kategori,
        data,
        color = 'red', #warna batang
        alpha = 0.25) #Tingkat transparansi

#Menambahkan garis garis kisi (grid)
plt.grid(linestyle='--',
         linewidth = 2,
         axis = 'y', #Menampilkan gris hanya pada sumbu y
         alpha = 0.50) #Tingkat transparansi grid

#Menambahkan label pada sumbu x dan y
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

#Menambahkan judul plot
plt.title('Contoh Bar Plot')

#Menampilkan plot
plt.show()
```

**Horizontal Bar Plot**

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
#Membuat horizontal bar plot
plt.barh(kategori, data)
# plt.barh(kategori[::-1], data[::-1])

#Menambahkan garis garis kisi (grid)
plt.grid(linestyle = '--',
         linewidth = 1,
         axis = 'x', #Menampilkan grid hanya pada sumbu x
         alpha = 0.75) #Tingkat transparansi grid

#Menambahkan label pada sumbu x dan y
plt.xlabel('Jumlah')
plt.ylabel('Kategori')

#Menampilkan judul plot
plt.title('Contoh Horizontal Bar Plot')

#Menampilakn plot
plt.show()
```

**Stacked Bar Plot**

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
#Membuat stacked bar plot dengan menumpuk data2 di atas data1
plt.bar(kategori, data1, label='Data 1')
#Parameter bottom digunakan untuk menentukan posisi data2 di atas data1
plt.bar(kategori, data2, label='Data 2', bottom=data1) 

#Menambahkan legenda untuk membedakan antara Data 1 dan Data 2
plt.legend()

#Menambahkan garis-garis kisi (grid)
plt.grid(linestyle='--', 
         linewidth=1, 
         axis='y', #Menampilkan grid hanya pada sumbu y
         alpha=0.75) #Tingkat transparansi grid

#Menambahkan label pada sumbu x dan y
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

#Menambahkan judul plot
plt.title('Contoh Stacked Bar Plot')

#Menampilkan plot
plt.show()
```

**Grouped Bar Plot**

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

#Array untuk deret bilangan 
x = np.arange(len(kategori))
x
```

```python
#Lebar setiap bar dalam grouped bar plot
width = 0.35

#Mendapatkan indeks untuk setiap kategori
x = np.arange(len(kategori))

#Membuat Grouped Bar Plot
plt.bar(x - width/2, data1, width, label='Data 1')
plt.bar(x + width/2, data2, width, label='Data 2')

#Menentukan label untuk setiap bar
plt.xticks(x, kategori)

#Menambahkan grid pada sumbu y dengan garis putus-putus
plt.grid(linestyle='--', 
         linewidth=1, 
         axis='y', 
         alpha=0.75)

#Menambahkan label pada sumbu x dan y
plt.xlabel('Kategori')
plt.ylabel('Jumlah')

#Menambahkan judul pada Grouped Bar Plot
plt.title('Contoh Grouped Bar Plot')

#Menampilkan legenda
plt.legend()

#Menampilkan Grouped Bar Plot
plt.show()
```

- Jika ingin membuat grouped bar plot, kita harus mengatur width (%) terlebih dahulu
- Semakin kecil width, semakin tipis bar yang dibuat, sehingga jarak antar bar semakin lebar
- Variabel x yang telah dibuat akan menjadi titik tengah kedua barplot, sehingga posisi data 1 yaitu x dikurangi oleh width dibagi dua, dan posisi data 2 yaitu X ditambah oleh width dibagi dua
- Parameter diatas kurang lebih memiliki arti (posisi, nilai data, lebar, dan nama label)
