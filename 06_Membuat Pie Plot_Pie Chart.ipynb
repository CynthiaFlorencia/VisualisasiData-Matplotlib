# **Membuat Pie Plot / Pie Chart**

Dalam sesi ini kita akan mempelajari cara membuat pie plot dengan Matplotlib

**Import Modules**

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

**Simple Pie Plot**

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
#Membuat pie plot
plt.pie(data,                    #Data yang akan di plot
        labels = kategori,       #Labels (kategori) untuk setiap sektor
        autopct = '%1.1f%%',     #Format tampilan persentase 
        startangle = 98)         #Sudut awal untuk rotasi

#Menambahkan judul pada pie plot
plt.title('Contoh Pie Plot')

#Untuk menampilkan pie plot
plt.show()
```

**Pengaturan Legend dan Color**

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#c03364', 'yellowgreen', 'skyblue']
```

```python
#Membuat pie plot dengan warna khusus
plt.pie(data,
        labels = kategori,
        colors = warna, #Mengganti warna setiap bagian sesuai urutan warna dalam daftar
        autopct = '%1.1f%%',
        startangle = 98)

#Menampilkan legenda
plt.legend()

#Untuk menambahkan judul pada pie plot
plt.title('Contoh Pie Plot')

#Untuk menampilkan pie plot
plt.show()
```

**Exploded Pie Plot**

```python
#Exploded pie plot berarti ada bagian yang akan ditarik keluar
#Semakin besar explode_var semakin jauh juga potongannya
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#c03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
```

```python
#Membuat Exploded Pie Plot
plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,  #Mengatur jarak sektor untuk memberikan efek "terpental" pada sektor tertentu
        autopct='%1.1f%%',
        startangle=90)

#Untuk menambahkan judul pada exploded pie plot
plt.title('Contoh Exploded Pie Plot')

#Untuk menampilkan exploded pie plot
plt.show()
```
