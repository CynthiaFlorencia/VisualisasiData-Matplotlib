**Membuat plotting sederhana**

```python
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')

#Fungsi show digunakan untuk menghilangkan output yang tidak perlu 
plt.show()
```

- Pemanggilan fungsi plot() dapat dilakukan dengan menyertakan hanya sebuah deret bilangan (list/array) ataupun dua buah deret bilangan (list/array)
- Bila pemanggilan hanya menyertakan sebuah deret bilangan, maka nilai pada deret bilangan tersebut akan dijadikan data point untuk nilai pada sumbu y, sedangkan nilai pada sumbu x akan secara otomatis dibuatkan sebuah deret bilangan terurut dengan nilai dimulai dari nol

```python
plt.plot([2, 3, 4, 7], [5, 4, 2, 16])
plt.show()
```

- Pada bagian deret bilangan pertama yaitu [2, 3, 4, 7] akan dijadikan data point untuk nilai pada sumbu x
- Pada deret bilangan kedua yaitu [5, 4, 2, 16] akan dijadikan data point untuk nilai pada sumbu y
- Deret deret tersebut dapat kita baca sebagai [(2,5), (3,4), (4,2), (7,16)]

**Pengaturan Format pada plot**

- Selain menyertakan dua buah deret bilangan sebagai parameter untuk sumbu x dan y, terdapat parameter ketiga yang bisa kita sertakan untuk mengatur warna dan jenis plotting
- Pengaturan parameter ketiga ini sepenuhnya mengadopsi gaya formatting pada MATLAB
- Nilai default untuk parameter ketiga ini adalah 'b-'
- Pemanggilan fungsi axis() menyertakan sebuah list [xmin, xmax, ymin, ymax]

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

- Parameter ketiga pengatur bentuk marker yang dimaksud adalah karakter o dan bukan 0
- Warna dari parameter ketiga dapat diubah menjadi r(red), g(green), dll
- Bentuk marker dari parameter ketiga dapat diubah menjadi 'o', '*', dll
- Contoh penggunaan parameter ketiga 'b-', 'g*', 'ro'
- Pada kasus ini, parameter ketiga yang digunakan adalah 'go' yang berarti bewarna 'green' dan berbentuk marker 'o'
- Fungsi axis() bertujuan untuk menentukan batas nilai dari setiap sumbu
- Rentang sumbu x dan sumbu y masing masing pada kasus ini adalah 0 - 6 dan 0 - 20

**Multi plot dalam satu pemanggilan fungsi plot()**

- Fungsi arange() memiliki parameter (min_value, max_value, step_value)
- Nilai 0. dan 5. menandakan bahwa mereka masing masing bernilai floating point yaitu '0f dan 5f'
- Nilai 0. dan 5. masing masing juga merepresentasikan nilai terendah dan nilai tertinggi

```python
t = np.arange(0., 5., 0.2)
t
```

- Untuk membuat plot dengan garis putus putus kita dapat menggunakan simbol -- 
- Untuk membuat plot dengan marker kotak kita dapat menggunakan simbol s (square)
- Untuk membuat plot dengan marker segitiga kita dapat menggunakan simbol ^

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

Kita juga dapat melakukan multi plot dalam satu pemanggilan fungsi plot()

```python
plt.plot(t, t, 'r--', 
         t, t**2, 'bs',
         t, t**3, 'g^')
plt.show()
```

**Plotting dengan keywords**
Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument (kwargs) seperti pada dictionary dan Pandas Data Frame

```python
#Key 'a' memiliki 50 values dari fungsi arange 
#Key 'c' memiliki 50 values dari random angka 0 sampai dengan 50 
#Key 'd' memiliki 50 values dari distribusi normal standar 
data = {'a':np.arange(50),
        'c':np.random.randint(0,50,50),
        'd':np.random.randn(50)
       }

#Menambahkan key'b' dengan values yang dihasilkan dari persamaan 
data['b'] = data['a'] + 10 * np.random.randn(50)

#Memodifikasi values pada key 'd'
#Mengubah values 'd' menjadi nilai absolut dan dikalikan dengan 100 
data['d'] = np.abs(data['d']) * 100

#Untuk menampilkan dictionary 'data'
data 
```

```python
#Membuat scatter plot dengan data dari keys 'a' dan 'b'
plt.scatter('a', 'b', data=data)

#Memberi label sumbu x dan sumbu y
plt.xlabel('entry a')
plt.ylabel('entry b')

#Untuk menampilkan scatter plot 
plt.show()
```

```python
#Membuat scatter plot dengan data dari keys a dan b, serta warna dari key c dan ukuran dari key d
plt.scatter('a', 'b', c='c', s='d', data=data)

#Memberi label sumbu x dan sumbu y 
plt.xlabel('entry a')
plt.ylabel('entry b')

#Untuk menampilkan scatter plot 
plt.show()
```
