# **Histogram**

Dalam sesi ini kita akan mempelajari cara membuat histogram dengan Matplotlib

**Import Modules**

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

**Sample Data**

```python
#Nilai 10000 pada fungsi rand() akan merepresentasikan jumlah data point yang dihasilkan
mu, sigma = 100, 15 #Nilai mean dan nilai standar deviation

#Menghasilkan 10.000 sampel dari normal distribution
x = mu + sigma *np.random.randn(10000) 

#Menampilkan data 
x
```

```python
#Menampilkan bentuk (shape) dari array x
x.shape
```

**Histogram dengan pyplot Style**

```python
#Membuat histogram
plt.hist(x,
         bins=50,
         facecolor='g',
         alpha=0.55)

#Menambahkan label pada sumbu x dan sumbu y
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

#Menambahkan judul pada histogram
plt.title('Contoh Histogram')

#Menambahkan teks pada plot
plt.text(45, 500, r'$\mu=100,\ \sigma=15$') #Menampilkan teks dengan rata rata dan deviasi standar

#Menampilkan grid pada plot
plt.grid()

#Menampilkan histogram
plt.show()
```

**Histogram dengan 00 Style**

```python
#Membuat subplot dengan objek axes (ax)
fig, ax = plt.subplots()

#Membuat histogram pada objek axes
ax.hist(x,
        bins=50,
        facecolor='r',
        alpha=0.25)

#Menentukan label pada sumbu x dan y
ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')

#Menentukan judul pada histogram
ax.set_title('Contoh Histogram')

#Menambahkan teks pada plot
ax.text(45, 500, r'$\mu=100,\ \sigma=15$')

#Menampilkan grid pada plot
ax.grid()

#Menampilkan Histogram
plt.show()
```
